## Introduction

### Dataset Description 

 Its dataset collects information from [100,000] medical appointments in Brazil and focuses on whether or not patients   show up for their appointments.

* ‘ScheduledDay’ tells us on what day the patient set up their appointment.
* ‘Neighborhood’ indicates the location of the hospital.
* ‘Scholarship’ indicates whether or not the patient is enrolled in Brasilian welfare program Bolsa Família.
* Be careful about the encoding of the last column: it says ‘No’ if the patient showed up to their appointment, and 
  ‘Yes’ if they did not show up.

```python
# Assign names
SHOW=df.No_show== 'No'
NO_SHOW=df.No_show== 'Yes'                                                          
```

### Comparison of those who attended and did not attend

```python
# Does gender affect the attendance rate
plt.figure(figsize=[16, 5])
df.Gender[SHOW].hist(alpha=0.6, color='blue', label='show')
df.Gender[NO_SHOW].hist(alpha=0.6, color='red', label='Noshow');
plt.title('who attended and who did not attend in terms of gender', size=15)
plt.xlabel('Gender',size=15)
plt.ylabel('The patients', size=15,)
plt.legend();
```

```python
# Attendance rate
print(df.Gender[show].value_counts())
```

```python
# Non attendance rate
print(df.Gender[Noshow].value_counts())
```

* The attendance rate of women was higher than that of men and the non-attendance rate was the same

## AGE

```python
# Does age affect attendance rate?
plt.figure(figsize=[16, 5])
df.Age[SHOW].hist(alpha=0.6, color='blue', label='show')
df.Age[NO_SHOW].hist(alpha=0.6, color='RED', label='Noshow');
plt.title('who attended and who did not attend in terms of Age', size=15)
plt.xlabel('Age',size=15)
plt.ylabel('The patients', size=15,)
plt.legend();
```

```python
# Attendance rate
print(df.Age[show].value_counts().sum())
```

```python
# Non attendance rate
print(df.Age[Noshow].value_counts().sum())
```

* The percentage of females is higher than that of males
* of [0:10], more than those who took an appointment, and the percentage of attendance was higher than absence
* In terms of [35:65] they were calculated, but at a lower rate than [0:10]
* The older the age over 65, the fewer attendees

## SMS RECEIVED

```python
# Does SMS received affect the attendance rate
plt.figure(figsize=[16, 5])
df.SMS_received[SHOW].hist(alpha=0.7, color='blue', label='show')
df.SMS_received[NO_SHOW].hist(alpha=0.7, color='orange', label='Noshow')
plt.title('who attended and who did not attend in terms of SMS received', size=15)
plt.xlabel('SMS received',size=15)
plt.ylabel('The patients', size=15,)
plt.legend();
```

* It seems to us that those who have received the messages are less present than those who have not

```python
# Are the received SMS messages sent differently from person to person?
plt.figure(figsize=[20, 7.5])
df[SHOW].groupby('Neighbourhood').SMS_received.mean().plot(kind='bar', alpha=0.5, color='blue', label='show')
df[NO_SHOW].groupby('Neighbourhood').SMS_received.mean().plot(kind='bar', alpha=0.7, color='orange', label='Noshow')
plt.title('who attended and who did not attend in terms of SMS_received', size=20)
plt.xlabel('SMS_received',size=15)
plt.ylabel('The patients', size=17,)
plt.legend();
```

* Only five neighborhoods responded to the letters
* This makes us want to know if the messages sent to all patients are the same


## NEIGHBOURHOOD

```python
# Does Neighbourhood affect the attendance rate
plt.figure(figsize=[20, 7.5])
df.Neighbourhood[SHOW].value_counts().plot(kind='bar' , color='blue', label='show')
df.Neighbourhood[NO_SHOW].value_counts().plot(kind='bar' , color='orange', label='Noshow')
plt.title('who attended and who did not attend in terms of Neighbourhood', size=20)
plt.xlabel('Neighbourhood',size=15)
plt.ylabel('The patients', size=17,)
plt.legend();
```

* It seems to us that the location strongly influences the rate of patient attendance or not
